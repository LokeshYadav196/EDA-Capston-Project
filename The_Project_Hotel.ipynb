# **Problem Statement**

### **Business Problem Overview: Understanding the Impact of Cancellations in the Hotel Industry**

#### In the hospitality industry, one of the most significant challenges hotels face is high cancellation rates which depands on customer behaviour. Frequent cancellations lead to revenue loss, lower occupancy rates, and inefficient resource allocation, ultimately impacting business profitability.

#### To address this issue, it is crucial to analyze the key factors influencing cancellations and develop data-driven strategies to mitigate them. This project aims to investigate the following aspects:

####**Cancellation Trends**: Identifying patterns in cancellations based on year, month, and seasonality to determine peak cancellation periods.
####**Customer Behaviour**: This factors impacts significantly to cancellation rates and also idenitifying it based on reserved room, assigned room, special request and many more.
#### **Guest Type Impact**: Analyzing whether first-time guests or repeated guests contribute more to cancellations.
#### **Lead Time Influence**: Evaluating how the number of days between booking and check-in (lead time) affects the likelihood of cancellations.
#### **Market Segment & Distribution Channel Analysis**: Understanding which booking channels and market segments have the highest cancellation rates.
#### **Price & ADR Correlation**: Examining whether higher Average Daily Rate (ADR) affects the probability of cancellations.

#### By identifying the root causes of cancellations, hotels can optimize their booking policies, pricing strategies, and customer engagement approaches to minimize cancellations and maximize revenue. This data-driven insight will help businesses enhance their operational efficiency and profitability.

### **Business Objective**

#### **Business Objective for Hotel Booking Analysis**
To analyze hotel booking data to identify key patterns and trends related to cancellations, no-shows, and deposit types. The goal is to optimize booking policies, minimize revenue loss, and improve customer retention by understanding factors influencing cancellations.

##### **Cancelation Status by Lead Time**

````python
most_cancelled_lead_time = df[df['is_canceled'] == 1]['lead_time'].value_counts().head(5)
most_cancelled_lead_time
````

````python
# Define bins for lead time
bins = [0, 50, 100, 200, 300, 400, 500, df['lead_time'].max()]
labels = ["0-50", "51-100", "101-200", "201-300", "301-400", "401-500", "500+"]

# Create a lead_time bucket column
df["lead_time_bucket"] = pd.cut(df["lead_time"], bins=bins, labels=labels, include_lowest=True)

# Grouping by hotel and lead time bucket
cancellation_by_lead_time_in_hotels = (df[df["is_canceled"] == 1].groupby(["hotel", "lead_time_bucket"])["is_canceled"].count().reset_index())

# Rename column for clarity
cancellation_by_lead_time_in_hotels.rename(columns={"is_canceled": "cancellation_count"}, inplace=True)

# Display the result
print(cancellation_by_lead_time_in_hotels)
````

````python
mean_lead_time = df['lead_time'].mean()
median_lead_time = df['lead_time'].median()
mode_lead_time = df['lead_time'].mode()[0] # [0] refers to position to show first indexing

print(f"Average (Mean) Lead Time: {mean_lead_time:.2f} days -mostly people prefer to book 80 days ago.")
print(f"Median Lead Time: {median_lead_time} days - half of the people book before this and same number of people book after this, it will help in showing demand trends")
print(f"Mode Lead Time: {mode_lead_time} days - maximum people are booking at the last moment. It comes in demand trends ")


max_lead_time = df['lead_time'].max()
min_lead_time = df['lead_time'].min()

print(f"Maximum Lead Time: {max_lead_time} days")
print(f"Minimum Lead Time: {min_lead_time} days")

cancelled_lead_time = df[df['is_canceled'] == 1]['lead_time']
not_cancelled_lead_time = df[df['is_canceled'] == 0]['lead_time']
print("Average Lead Time (Cancelled Bookings):", cancelled_lead_time.mean())
print("Average Lead Time (Confirmed):", not_cancelled_lead_time.mean())
````
