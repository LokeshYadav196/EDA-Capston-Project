# **Project Summary: Yes Bank Stock Price Regression Predicition and Analysis using Machine Learning**

This project focuses on analyzing and forecasting the stock price of Yes Bank using a regression-based machine learning approach. The dataset comprises historical stock price data, including daily values of **Open**, **High**, **Low**, **Close**, and over several years. The objective is to explore data, engineer relevant features, identify key insights, and build a predictive model for the **next day's closing price**.

---

### **Understanding the Dataset**

The dataset initially contained **185 rows and 5 columns**. We verified that there were **no null or duplicate values** at the beginning. Each record represented one trading day’s activity. The primary goal was to predict the **next day’s closing price**, so we created a new column named:

```python
df['Target'] = df['Close'].shift(-1)
```

This column represents the **target variable** (y), shifting the ‘Close’ price by one day ahead — a common strategy in time series forecasting when predicting future prices.

---

### **Feature Engineering**

To enhance the model’s predictive power, we created several new columns:

* **Month** and **Year**: Extracted from the Date column to analyze seasonal and yearly trends.
* **Volatility** = High – Low: Shows the price fluctuation during the day.
* **Price Movement** = Close – Open: Indicates whether the stock went up or down intraday.
* **Pct Return** = (Close – Open) / Open: Shows the percentage return for each day.
* **SMA\_5**, **SMA\_10**: 5-day and 10-day Simple Moving Averages to capture short-term price trends.
* **RSI\_14**: Relative Strength Index with 14-period window — used to identify overbought or oversold conditions.

These features are crucial for revealing hidden trends and improving model performance.

**Note:** After feature engineering, **some null values were generated** (e.g., from SMA and RSI computations), so we **removed those rows** to clean the data again.

### **Feature Selection & Multicollinearity Handling**

During analysis, we identified **multicollinearity** between some features (like Open, High, Low). To prevent the model from becoming unstable or biased, we filtered out highly collinear features before training.

This ensured that our regression models would be **more reliable, less overfitted**, and generalize better on unseen data.

## 3. ***Data Wrangling***

In the initial inspection of the dataset, it was observed that the primary columns — **Date, Open, High, Low, and Close** — contained **no missing values** and **no duplicate records**. As a result, **traditional data cleaning steps** such as null handling and duplicate removal were not required for these columns.

At the beginning of the analysis, the Date column was successfully converted into DateTime format, which allowed for efficient time-based feature extraction and analysis.


However, to enhance the dataset for better **analysis and modeling**, several new columns were derived through **feature engineering**. This included extracting time-based features like **Month** and **Year** from the Date column, and creating financial indicators such as:

* **Volatility** (`High - Low`)
* **Price Movement** (`Close - Open`)
* **Percentage Return** (`(Close - Open) / Open`)
* **Simple Moving Averages** (`SMA_5`, `SMA_10`)

Although these transformations were primarily feature engineering steps, they **implicitly contributed to data cleaning** by enriching the dataset and highlighting any inconsistencies, if present, in the original pricing trends.

---


### **Feature Engineering for EDA and Visulaization Purposes**

---



````python
# Write your code to make your dataset analysis ready

df['Month'] = df['Date'].dt.month #extract Month Column

import calendar

df['Month'] = df['Month'].astype(int)  # ensure it's numeric
df['Month'] = df['Month'].apply(lambda x: calendar.month_name[x])
month_order = list(calendar.month_name)[1:]  # Jan to Dec
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)


df['Year'] = df['Date'].dt.year #extract Year Column

df['Volatility'] = df['High'] - df['Low'] # create Price Range Column {Stock ki Volatility dekhne ke liye}

df['Price Movement'] = df['Close'] - df['Open'] # Creat Change Columnv {Kitna uper ya neeche gaya stock}

df['Pct Return'] = ((df['Close'] - df['Open'])/ df['Open']) * 100 # Perentage Change Column {Relative Price Movement by Percemtage}

df['SMA_5'] = df['Close'].rolling(window=5).mean() # Simple Moving Average

df['SMA_10'] = df['Close'].rolling(window=10).mean() # Simple Moving Average

df.head()
````

````python
df.isnull().sum()
````

````python
df = df.dropna()
df.info()
````

````python
# Calculate IQR for Close price to detect outliers
Q1_close = df['Close'].quantile(0.25)
Q3_close = df['Close'].quantile(0.75)
IQR_close = Q3_close - Q1_close

lower_close = Q1_close - 1.5 * IQR_close
upper_close = Q3_close + 1.5 * IQR_close

# Label outliers based on Close price
df['Outlier'] = df['Close'].apply(lambda x: 'Peak' if x > upper_close else ('Trough' if x < lower_close else 'Normal'))

# Plot the Close price with outliers
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='gray', alpha=0.6)
plt.scatter(df[df['Outlier'] == 'Peak']['Date'], df[df['Outlier'] == 'Peak']['Close'], color='green', label='Peak (High Outlier)', zorder=5)
plt.scatter(df[df['Outlier'] == 'Trough']['Date'], df[df['Outlier'] == 'Trough']['Close'], color='red', label='Trough (Low Outlier)', zorder=5)

plt.axhline(y=upper_close, color='green', linestyle='--', alpha=0.5)
plt.axhline(y=lower_close, color='red', linestyle='--', alpha=0.5)

plt.title('📈 Close Price Outliers: Peaks and Troughs')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## **7. ML Model Implementation**

#### **Model Selection**

````python
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'KNN Regressor': KNeighborsRegressor()
}
````

#### **Model Evaluation Metrics**


````python
results = []

n = X_test_scaled.shape[0]  # number of samples in test
k = X_test_scaled.shape[1]  # number of features

def adjusted_r2(r2, n, k):
    return 1 - ((1 - r2) * (n - 1) / (n - k - 1))

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_test, y_pred)
    adj_r2 = adjusted_r2(r2, n, k)

    results.append({
        'Model': name,
        'MAE': round(mae, 2),
        'MSE': round(mse, 2),
        'RMSE': round(rmse, 2),
        'R2 Score': round(r2, 4),
        'Adjusted R2': round(adj_r2, 4)
    })

results_df = pd.DataFrame(results)
````

   * **RMSE (Root Mean Squared Error)**

   * **Adjusted R² Score** (to penalize unnecessary features)
   * The **Adjusted R²** was manually calculated using the formula:

   $$
   \text{Adjusted R²} = 1 - \left( \frac{(1 - R^2) \cdot (n - 1)}{n - k - 1} \right)
   $$

   where `n` is the number of test samples and `k` is the number of features.
